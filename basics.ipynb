```python
# ======================================
# ESEMPI COMPLETI SUI METODI DELLE LISTE
# ======================================

import random

print("\n=== LISTA DI PARTENZA ===")
myList = ['a','b','c','d','e','f']
print("Lista iniziale:", myList)

# 1Ô∏è‚É£ append()
print("\n--- append() -> aggiunge un elemento in fondo ---")
myList.append('g')
print("Dopo append('g'):", myList)

# 2Ô∏è‚É£ insert()
print("\n--- insert() -> inserisce un elemento in una posizione specifica ---")
myList.insert(2, 'X')   # inserisco X alla posizione indice 2
print("Dopo insert(2, 'X'):", myList)

# 3Ô∏è‚É£ extend()
print("\n--- extend() -> aggiunge molti elementi alla lista ---")
myList.extend(['y', 'z'])
print("Dopo extend(['y','z']):", myList)

# 4Ô∏è‚É£ copy()
print("\n--- copy() -> crea una copia indipendente della lista originale ---")
copiedList = myList.copy()
print("Copia della lista:", copiedList)

# 5Ô∏è‚É£ pop()
print("\n--- pop() -> rimuove e ritorna un elemento (default ultimo) ---")
popped_item = myList.pop()
print("Elemento rimosso:", popped_item)
print("Lista dopo pop():", myList)

# rimozione con indice
popped_item2 = myList.pop(1)
print("Elemento rimosso in indice 1:", popped_item2)
print("Lista dopo pop(1):", myList)

# 6Ô∏è‚É£ remove()
print("\n--- remove() -> rimuove la prima ricorrenza di un elemento ---")
copiedList.remove('X')   # rimuovo la X dalla lista copiata
print("Lista copiata dopo remove('X'):", copiedList)

# 7Ô∏è‚É£ reverse()
print("\n--- reverse() -> inverte l‚Äôordine degli elementi ---")
tempList = copiedList.copy()
tempList.reverse()
print("Lista invertita con reverse():", tempList)

# 9Ô∏è‚É£ count()
print("\n--- count() -> conta quante volte un elemento appare nella lista ---")
print("Conteggio della lettera 'a':", myList.count('a'))

# üîü index()
print("\n--- index() -> restituisce l‚Äôindice della prima occorrenza ---")
el = random.choice(copiedList)
if el in myList:
    print(f"Indice di {el}:", myList.index(el))
else:
    print("el non si trova pi√π nella lista!")

# üîÑ clear()
print("\n--- clear() -> svuota completamente la lista ---")
temp = ['uno', 'due', 'tre']
print("Prima:", temp)
temp.clear()
print("Dopo clear():", temp)

```

```python
# Example 1: Sort the list in reverse order
technology = ['Java','Hadoop','Spark','Pandas','Pyspark','NumPy','Hyperion']
sorted_list = sorted(technology)
print(sorted_list)

# Example 2: Sort list in reverse order
technology.sort(
    reverse=True #ordine opposto
    )
print(technology)

# Example 3: Sort list in reverse order by length
technology.sort(
    reverse = True, #ordine opposto
    key = len #function for the criterion of sorting
    )
print(technology)

technology.sort(
    key = lambda x: x[::-1] #function for the criterion of sorting
    )
print(technology)
```

Defoult dictionaries how to countwords in a document

```python
from collections import defaultdict

# ======================================
# 1Ô∏è‚É£ CREAZIONE DOCUMENTO DI TESTO
# ======================================
filename = "documento.txt"

contenuto = """apple pear apple orange pear apple banana pear apple
banana kiwi kiwi apple"""

# scrivo il contenuto nel file
with open(filename, "w") as f:
    f.write(contenuto)

# leggo il documento e ottengo le parole in lista
with open(filename, "r") as f:
    document = f.read().split()

print("Documento:", document)
print("=" * 50)


# ======================================
# 2Ô∏è‚É£ DIVERSI APPROCCI (FUNZIONI)
# ======================================

def count_normal(doc):
    """Usa dict + if check"""
    counts = {}
    for w in doc:
        if w in counts:
            counts[w] += 1
        else:
            counts[w] = 1
    return counts

def count_try_except(doc):
    """EAFP (forgiveness > permission)"""
    counts = {}
    for w in doc:
        try:
            counts[w] += 1
        except KeyError:
            counts[w] = 1
    return counts

def count_get(doc):
    """Usa .get() per valore default"""
    counts = {}
    for w in doc:
        counts[w] = counts.get(w, 0) + 1
    return counts

def count_defaultdict(doc):
    """Usa defaultdict (pi√π pulito)"""
    counts = defaultdict(int)  # int() ‚Üí 0
    for w in doc:
        counts[w] += 1
    return dict(counts)


# ======================================
# 3Ô∏è‚É£ MOSTRO I RISULTATI
# ======================================

print("Normal dict:", count_normal(document))
print("Try/Except:", count_try_except(document))
print("get():", count_get(document))
print("defaultdict:", count_defaultdict(document))
```

```python
# --- Other defaultdict types ---
dd_list = defaultdict(list)       # list() ‚Üí []
dd_list[2].append(1)

dd_dict = defaultdict(dict)       # dict() ‚Üí {}
dd_dict["Joel"]["City"] = "Seattle"
dd_dict["Joel"]["Age"]= 23
dd_dict["Joel"]["City"] = "Seattle"
dd_dict["Joel"]["Age"]= 23

dd_pair = defaultdict(lambda: [0, 0]) 
dd_pair[2][1] = 1

print("list example:", dict(dd_list))        # {2: [1]}
print("dict example:", dict(dd_dict))        # {'Joel': {'City': 'Seattle'}}
print("pair example:", dict(dd_pair))        # {2: [0,1]}
```
