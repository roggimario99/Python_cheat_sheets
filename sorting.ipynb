Task B ‚Äî Sort a list of strings alphabetically

Given:

names = ["zoe", "Alex", "bob", "charlie", "anna"]


- Sort alphabetically case-insensitive
- Then try sorting with case sensitivity preserved (uppercase first)

```python
names = ["zoe", "alex", "bob", "charlie", "Anna", "Ab",  "ac","Ac", "ab"]
names_sort_case_sens = sorted(names) #case sensitive: upper case first
names_sort_case_insens = sorted(names, key = lambda x: x.lower()) #case insensitive

print("case sensitive: ", names_sort_case_sens)
print("case insensitive: ", names_sort_case_insens)
```

Task D ‚Äî Sort a dictionary by value, key, and by custom rule

```python
scores = {
    "Alice": 82,
    "Bob": 95,
    "Charlie": 70,
    "David": 95,
    "Eve": 88,
    "Evet": 88,
    "Evella": 88,
    "Evuccia": 88
}

scores_sorted = dict(sorted(scores.items())) #sort by key (defoult first element of the tuple)
print(scores_sorted)

scores_sorted = dict(sorted(scores.items(), key = lambda x: x[1])) #sort by value 
print(scores_sorted)

print("-"*30)
scores_sorted = dict(sorted(scores.items(), key = lambda x: (x[1], len(x[0])))) #sort by value, if equal by increasing name lenght
print(scores_sorted)

scores_sorted = dict(sorted(scores.items(), key = lambda x: (x[1], -len(x[0])))) #sort by value, if equal by decreasing name lenght 
print(scores_sorted)
```

üî¥ Exercise 3 ‚Äî Advanced Sorting with Custom Functions
Task E ‚Äî Sort by computed/complex criteria

Given:

sentences = [
    "Python is great!",
    "Sort me by my length please",
    "Hi",
    "I love sorting challenges",
]


Sort the list by:

1Ô∏è‚É£ Length of the sentence
2Ô∏è‚É£ Number of vowels inside the sentence
3Ô∏è‚É£ Word count
4Ô∏è‚É£ In descending order of vowel percentage = (#vowels / length)

üëâ Create one function:

def sort_metric(sentence):
    ...


Then call:

sorted(sentences, key=sort_metric, reverse=True)

```python
sentences = [
    "Python is great!",
    "Sort me by my length please",
    "Hi",
    "I love sorting challenges",
    "a b c d e f g"
]

sentences_sorted = sorted(sentences, key = len) #sort by key (defoult first element of the tuple)
print(sentences_sorted)

def sort_num_words(sentence):
    return len(sentence.split(" "))

sentences_sorted = sorted(sentences, key = sort_num_words) #sort by key (defoult first element of the tuple)
print(sentences_sorted)
```

```python
def sort_num_vowels(sentence):
    n_vowels = 0
    vowels = ["a", "e", "i", "o", "u"]
    for let in sentence:
        if let in vowels:
            n_vowels += 1
        
    return n_vowels

sentences_sorted = sorted(sentences, key = sort_num_vowels) #sort by key (defoult first element of the tuple)
print(sentences_sorted)


def sort_num_vowels_ratio(sentence):
    n_vowels = 0
    vowels = ["a", "e", "i", "o", "u"]
    for let in sentence:
        if let in vowels:
            n_vowels += 1
        
    return n_vowels / len (sentence)

sentences_sorted = sorted(sentences, key = sort_num_vowels_ratio) #sort by key (defoult first element of the tuple)
print(sentences_sorted)
print( dict([ (s, round(sort_num_vowels_ratio(s),2) ) for s in sentences_sorted]) )
```
